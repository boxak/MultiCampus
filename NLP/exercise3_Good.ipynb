```r
library('dplyr')
library('xlsx')
library('writexl')
library('stringr')
```

```r
data2 <- read.xlsx2("DATA_02.xls",sheetIndex=1,startRow=1,header=TRUE)
head(data2)
```

```r
names(data2)
```

```r
data2[,"어휘"] <- NULL
data2[,"구성.단위"] <- NULL
data2[,"고유어.여부"] <- NULL
data2[,"로마자"] <- NULL
data2[,"로마자.글자"] <- NULL
data2[,"로마자.소리"] <- NULL
data2[,"원어"] <- NULL
data2[,"어원"] <- NULL
data2[,"발음"] <- NULL
data2[,"활용"] <- NULL
data2[,"검색용.이형태"] <- NULL
data2[,"방언.지역"] <- NULL
data2[,"문형"] <- NULL
data2[,"문법"] <- NULL
data2[,"분류"] <- NULL
data2[,"용례"] <- NULL
data2[,"전문.분야"] <- NULL
data2[,"속담"] <- NULL
data2[,"관용구"] <- NULL
data2[,"대역어"] <- NULL
data2[,"생물.분류군.정보"] <- NULL
data2[,"역사.정보"] <- NULL
data2[,"규범.정보"] <- NULL
data2[,"다중.매체.멀티미디어..정보"] <- NULL
data2[,"sense_no"] <- NULL
data2[,"word_no"] <- NULL
data2[,"다의어.번호"] <- NULL
data2[,"다의어.순번"] <- NULL
head(data2)
```

```r
뜻풀이.음절 <- vector()
size <- nrow(data2)
for(i in 1:size){
    뜻풀이.음절 <- c(뜻풀이.음절,str_length(data2[i,3])-str_count(data2[i,3],' ')-str_count(data2[i,3],',')-str_count(data2[i,3],'\\(')-str_count(data2[i,3],'\\)')-str_count(data2[i,3],fixed("."))-str_count(data2[i,3],'‘')-str_count(data2[i,3],'’'))
}
cbind(data2,뜻풀이.음절)
writexl::write_xlsx(data2, path = "RefinedData002.xlsx")
```

```r
뜻풀이.어절 <- vector()
for(i in 1:size){
    뜻풀이.어절 <- c(뜻풀이.어절,str_count(data2[i,3],' ')+1)
}
cbind(data2,뜻풀이.어절)
head(data2)
```

```r
writexl::write_xlsx(cbind(data2,뜻풀이.음절,뜻풀이.어절), path = "RefinedData002.xlsx")
```
